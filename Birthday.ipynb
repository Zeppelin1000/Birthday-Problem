```python
import datetime
import random
import matplotlib.pyplot as plt 
import math
```

```python
people_no=[]
probability = []
theo_probs=[]
for k in range(23,45):
    trials = 2000
    successes = 0
    for j in range(1,trials):
        date_list =[]
        people = k
        
        for i in range (0,k):
            start_date = datetime.date(2020, 1, 1)
            end_date = datetime.date(2023, 12, 31)
            time_between_dates = end_date - start_date
            days_between_dates = time_between_dates.days
            random_number_of_days = random.randrange(days_between_dates)
            #days_between_dates = time_between_dates.days
            #random_number_of_days = random.randrange(days_between_dates)
            random_date = start_date + datetime.timedelta(days=random_number_of_days)
            date = str(random_date.month) + '/' + str(random_date.day)
            date_list.append(date)
        #print(date_list)
        my_list = list(set(date_list))
        if len(date_list) != len(my_list):
            successes +=1
        #print('not equal')
    #print('365-k')    
    #print(365-k+1)    
    theoretical = 1- (math.factorial(365)/math.factorial(365-k))*(1/365)**k
    theo_probs.append(theoretical)
    people_no.append(k)
    probability.append(successes/trials)
print(theo_probs)    
plt.plot(people_no, probability,label='Experimental') 
plt.plot(people_no, theo_probs,label='Theoretical')     
# naming the x axis 
plt.xlabel('No of People') 
# naming the y axis 
plt.ylabel('Probability') 
    
# giving a title to my graph 
plt.title('The Birthday Problem') 
plt.legend(loc="upper left")   
# function to show the plot 
plt.show() 
#plt.savefig('birthday')
```
